```python
import matplotlib.pyplot as plt
import numpy as np
import glob, os, dill
import anndata
import wot
import pandas as pd
import pegasus as pg
PLT_CELL = 3.75
```

```python
def KL(a, b):
    eps = 1e-16
    return ((a+eps)*np.log((a+eps)/(b+eps))).sum()

def TV(a, b):
    return (np.abs(a - b)*0.5).sum()
```

```python
# set random seed 
np.random.seed(0)
adata = anndata.read_h5ad("/zfs/users/syz/syz/sde_couplings/data_repr.h5ad")
cell_sets = wot.io.read_sets("cell_sets.gmt")
obs_celltype = pd.DataFrame(cell_sets.X, dtype = bool, columns = cell_sets.var.index, index = cell_sets.obs.index)
idx_batch = pd.Index.intersection(cell_sets.obs.index, adata.obs.index)
adata_batch = adata[idx_batch, ]
adata_batch.obs = pd.concat([adata_batch.obs, obs_celltype.loc[idx_batch, ]], axis = 1)

adata_batch = adata_batch[adata_batch.obs.day > 4, ]
del adata_batch.uns
```

```python
adata_ts = adata_batch[~np.isnan(adata_batch.obs.day), :] 
pg.pca(adata_ts, n_components = 10, features = None)
t_map = np.array(adata_ts.obs.day.unique())
t_map = t_map[t_map >= t_initial]
t_map = t_map[t_map < t_final]
t_map.sort()
```

```python
os.chdir("/zfs/users/syz/syz/reprogramming_batch")
lamda_vals = []
dir_names = []
for fname in glob.glob("run_*_dirichlet5"):
    lamda_vals += [np.float(fname.split("_")[1]), ]
    dir_names += [fname, ]
lamda_vals = np.array(lamda_vals)
```

```python
kl_summ = []
kl_ratio = []
kl_null_ratio = []

def kl_summary(out):
    props = out['props']
    props_gwot = out['props_gwot']
    props_subsamp = out['props_subsamp']
    p_null = np.array(props_subsamp).mean(0)
    kl_gwot = np.array([   KL(np.array(props[i]), np.array(props_gwot[i])) for i in range(len(props))])
    kl_subsamp = np.array([KL(np.array(props[i]), np.array(props_subsamp[i])) for i in range(len(props))])
    kl_null = np.array([   KL(np.array(props[i]), p_null) for i in range(len(props))])
    return kl_gwot, kl_subsamp, kl_null

for dirname in dir_names:
    os.chdir(dirname)
    out_all = []
    for fname in glob.glob("*.out"):
        with open(fname, "rb") as f:
            out_all += [dill.load(f), ]
    out_all = out_all[:25]
    # 
    kl_summ += [np.array([np.array(kl_summary(out)) for out in out_all]), ]
    kl_ratio += [kl_summ[-1][:, 0, :]/kl_summ[-1][:, 1, :], ]
    kl_null_ratio += [kl_summ[-1][:, 2, :]/kl_summ[-1][:, 1, :], ]
    os.chdir("..")
```

```python
def set_violin_color(v, c, alpha = None):
    for k in v.keys():
        if k == "bodies":
            for pc in v[k]:
                pc.set_facecolor(c)
                pc.set_edgecolor(c)
                if alpha is not None:
                    pc.set_alpha(alpha)
        else:
            pc = v[k]
            pc.set_edgecolor(c)
            if alpha is not None:
                pc.set_alpha(alpha)
    return 
```

```python
k = np.where(lamda_vals == 0.0025)[0][0]
plt.figure(figsize = (3*PLT_CELL, PLT_CELL))
plt.subplot(1, 2, 1)
viol = plt.violinplot(kl_summ[k][:, 0, :], showmeans = True, positions = np.arange(len(t_map)), widths = 0.3);
set_violin_color(viol, "blue", alpha = None)
viol = plt.violinplot(kl_summ[k][:, 1, :], showmeans = True, positions = np.arange(len(t_map)) - 0.25, widths = 0.3);
set_violin_color(viol, "red", alpha = None)
viol = plt.violinplot(kl_summ[k][:, 2, :], showmeans = True, positions = np.arange(len(t_map)) + 0.25, widths = 0.3);
set_violin_color(viol, "green", alpha = None)
plt.yscale("log")

leg = plt.legend(['gWOT', 'raw', 'null'], loc = "lower right")
leg.legendHandles[0].set_color("blue")
leg.legendHandles[0].set_alpha(0.5)
leg.legendHandles[1].set_color("red")
leg.legendHandles[1].set_alpha(0.5)
leg.legendHandles[2].set_color("green")
leg.legendHandles[2].set_alpha(0.5)
plt.title("KL($p$ | $\\cdot$)")
plt.xticks(np.arange(len(t_map)), t_map, rotation = 45);
plt.xlabel("day")
# plt.title("$\\lambda$ = %0.4f" % lamda_vals[k])

plt.subplot(1, 2, 2)
viol = plt.violinplot(kl_ratio[k], showmeans = True, positions = np.arange(len(t_map)) - 0.125, widths = 0.3);
set_violin_color(viol, "blue", alpha = None)
viol = plt.violinplot(kl_null_ratio[k], showmeans = True, positions = np.arange(len(t_map)) + 0.125, widths = 0.3);
set_violin_color(viol, "green", alpha = None)
plt.yscale("log")
plt.hlines(1, xmin = -1, xmax = len(t_map), color = "red", linestyle = "dashed")
plt.title("KL($p$ | $\\cdot$)/KL($p$ | $\\hat{p}$)")
# plt.title("$\\lambda$ = %0.4f" % lamda_vals[k])
plt.xticks(np.arange(len(t_map)), t_map, rotation = 45);
plt.xlabel("day")
leg = plt.legend(['gWOT', 'null'], loc = "lower right")
leg.legendHandles[0].set_color("blue")
leg.legendHandles[0].set_alpha(0.5)
leg.legendHandles[1].set_color("green")
leg.legendHandles[1].set_alpha(0.5)
plt.savefig("prop_ratio_kl.pdf", bbox_inches='tight')
```
